```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv('WineQT.csv')
df.head()
```

```python

plt.style.use('seaborn-v0_8-whitegrid')

# Create a histogram of the 'quality' column
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='quality', bins=6, kde=True, legend=True)

# Add title and labels
plt.title('Distribution of Wine Quality')
plt.xlabel('Quality')
plt.ylabel('Count')

# Add a legend (histplot automatically adds a legend for kde if set to True)
plt.legend(['Density Estimation', 'Count'])

# Show the plot
plt.show()
```

```python
# Create a scatter plot of alcohol vs quality
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='alcohol', y='quality', alpha=0.5)

# Add title and labels
plt.title('Alcohol vs Wine Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')

# Show the plot
plt.show()
```

```python
# Set the style for the plot
plt.style.use('seaborn-v0_8-whitegrid')

# Create box plots for each feature against quality
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
            'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

plt.figure(figsize=(12, 8))
for i, feature in enumerate(features):
    plt.subplot(3, 4, i + 1)
    sns.boxplot(data=df, x='quality', y=feature)
    plt.title(f'{feature} vs Quality')
    plt.xlabel('Quality')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
# Set the style for the plot
plt.style.use('seaborn-v0_8-whitegrid')

# Create violin plots for each feature against quality
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
            'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

plt.figure(figsize=(12, 8))
for i, feature in enumerate(features):
    plt.subplot(3, 4, i + 1)
    sns.violinplot(data=df, x='quality', y=feature, inner='quartile')
    plt.title(f'{feature} vs Quality')
    plt.xlabel('Quality')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```

```python
# Set the style for the plot
plt.style.use('seaborn-v0_8-whitegrid')

# Create a line plot of alcohol content
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['alcohol'], color='indianred', linestyle='-', linewidth=1)

# Add title and labels
plt.title('Alcohol Content Across Samples')
plt.xlabel('Sample Index')
plt.ylabel('Alcohol Content')

# Show the plot
plt.show()
```
