```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
# prompt: add bar plot with styles and legend

import matplotlib.pyplot as plt
# Count the occurrences of each class
class_counts = df['Class'].value_counts()

# Create a bar plot
plt.figure(figsize=(6, 4))
bar_plot = sns.barplot(x=class_counts.index, y=class_counts.values, palette='viridis')

# Add labels and title
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Distribution of Classes in Credit Card Data')

# Add legend (although for a simple bar plot with two bars, it might be redundant, but demonstrated for completeness)
# Create custom legend handles
legend_handles = [plt.Rectangle((0,0),1,1, color=bar_plot.patches[0].get_facecolor()),
                  plt.Rectangle((0,0),1,1, color=bar_plot.patches[1].get_facecolor())]
plt.legend(legend_handles, ['Class 0 (Non-Fraud)', 'Class 1 (Fraud)'])

# Add value labels on top of bars
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='center', xytext=(0, 5), textcoords='offset points')

# Show the plot
plt.show()
```

```python
# prompt: add box plot with legendn

import matplotlib.pyplot as plt
# Assuming 'Amount' is a relevant numerical feature for box plot
plt.figure(figsize=(8, 6))
box_plot = sns.boxplot(x='Class', y='Amount', data=df, palette='viridis')

# Add labels and title
plt.xlabel('Class')
plt.ylabel('Transaction Amount')
plt.title('Box Plot of Transaction Amount by Class')

# Show the plot
plt.show()
```

```python
# prompt: pie char in diffrent color and add legend

import matplotlib.pyplot as plt
# Assuming 'Class' is the categorical variable and 'Amount' or a count is used for the pie chart
# Let's use the 'Class' distribution for the pie chart as a typical example

# Count the occurrences of each class
class_counts = df['Class'].value_counts()

# Define colors for the slices
colors = ['lightblue', 'lightcoral'] # Example colors, choose based on preference

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)

# Add a title
plt.title('Distribution of Classes')

# Add a legend
plt.legend(labels=['Class 0 (Non-Fraud)', 'Class 1 (Fraud)'], loc='best')

# Ensure the pie chart is circular
plt.axis('equal')

# Show the plot
plt.show()
```

```python
# Histogram of 'Amount'
plt.figure(figsize=(10, 6))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Transaction Amount')
plt.show()

# Histogram of 'Amount' by 'Class'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Amount', hue='Class', bins=50, kde=True)
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Distribution of Transaction Amount by Class')
plt.show()
```

```python
# Violin plot of 'Amount' by 'Class'
plt.figure(figsize=(8, 6))
sns.violinplot(x='Class', y='Amount', data=df, palette='viridis')
plt.xlabel('Class')
plt.ylabel('Transaction Amount')
plt.title('Violin Plot of Transaction Amount by Class')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False) # annot=False to avoid clutter with many features
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
# Scatter plot of 'Time' vs 'Amount' colored by 'Class'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=df, alpha=0.6)
plt.xlabel('Time (seconds)')
plt.ylabel('Transaction Amount')
plt.title('Scatter Plot of Transaction Time vs. Amount by Class')
plt.show()
```
